```python
import pandas as pd
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from glob import glob
import os
from PIL import Image
import cv2
from tqdm import tqdm
```

```python
df = pd.read_csv('PATH_TO_DATA/train_256_bbs_final.csv')
df
```

```python
# Select images without bounding box only

df_noBox = df[df['x1'] == -1]
df_noBox
```

```python
for img in tqdm(df_noBox['challenge_id']):
    
    path = glob(os.path.join('PATH_TO_DATA','AIROGS_train','*',img+'*'))[-1]    
    new_path ='PATH_TO_SAVE_CROPPED_IMAGES'

    os.makedirs(new_path.replace(os.path.basename(new_path),''),exist_ok=True)
    
    im = plt.imread(path) 
    w,h,_ = im.shape
    
    transform = transforms.Compose([
      transforms.ToTensor(),                            
      transforms.CenterCrop(w*0.85),
      transforms.Resize((512, 512))
    ])

    _img = transform(im)
    _img = (_img.moveaxis(0, -1).numpy() * 255).astype('uint8')
    
    assert cv2.imwrite(f'PATH_TO_SAVE_CROPPED_IMAGES/{img}.jpg', cv2.cvtColor(_img, cv2.COLOR_RGB2BGR))
```
